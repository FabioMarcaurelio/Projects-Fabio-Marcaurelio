```python
#Quanti biglietti venduti per ogni classe?

pclass_df = df.groupby("Pclass").count()["PassengerId"].to_frame(name="Biglietti venduti") #cosÃ¬ creiamo un dataframe 
pclass_df
```

```python
pclass_df.plot(kind="bar") #plot su tipologia classe
```

```python
#Quanti sopravvissuti per ogni classe?

df_survived= df[["Pclass","Survived"]].groupby('Pclass').mean()
df_survived #piÃ¹ sopravvissuti in prima classe
```

```python
#reset_index():

#This function resets and provides the new index to the grouped by dataframe and makes them a proper dataframe structure.

df_survived_2= df[["Pclass","Survived"]].groupby('Pclass').mean().reset_index()
df_survived_2
```

```python
df_survived.plot(kind='bar') #plot su dataframe creato poco fa
```

```python
df[['Survived', 'Sex']].groupby('Sex').mean().plot(kind='bar') #mean su sesso dei sopravvissui
```

```python
df[['Survived', 'Pclass']].groupby('Pclass').mean().plot(kind='bar')
```

```python
survivors_df = df[['Survived', 'Sex']].groupby('Sex').sum() #SUM VALUES: 0 (NOT SURVIVED) , 1 (SURVIVED)
survivors_df['Total'] = df[['Survived', 'Sex']].groupby('Sex').count() #COUNT TOTAL NUMBER OF PEOPLE PER CLASS (MALE,FEMALE)
survivors_df.plot(kind='bar')
```

```python
survivors_df['%']= (survivors_df['Survived']*100)/survivors_df['Total']
survivors_df
```

```python
children_df = df[df['Age'] <= 17] #BOOLEAN INDEXING bambini minore di 17 anni
children_df.head()
```

```python
total_children= children_df.shape[0] #number of rows per stimare numero bambini
total_children
```

```python
children_survived= children_df['Survived'].sum() #somma valori zero e uno
children_survived
```

```python
adults_df= df[df['Age'] > 17]
adults_df= adults_df.rename(columns={'Sex': 'People'}) #sex cambia nome in people

survivors_df = adults_df[['Survived', 'People']].groupby('People').sum() #SUM VALUES: 0 (NOT SURVIVED) , 1 (SURVIVED)
survivors_df['Total'] = adults_df[['Survived', 'People']].groupby('People').count() #COUNT TOTAL NUMBER OF PEOPLE PER CLASS (MALE,FEMALE)
survivors_df['%']= (survivors_df['Survived']*100)/survivors_df['Total']
survivors_df
```
